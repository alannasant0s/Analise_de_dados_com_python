II) **Criando um cliente no bigquery. As tabelas utilizadas neste projeto foram criadas com a biblioteca Faker,a qual gerou 1000 linhas ficticias para cada tabela.**

```python
client = bigquery.Client(project='dados-400119')
```

# **Listar datasets do projeto**




```python
datasets = list(client.list_datasets())

if datasets:
    print("Datasets no projeto 'dados-400119':")
    for dataset in datasets:
        print(f"- {dataset.dataset_id}")
else:
    print("Nenhum dataset encontrado.")
```

# Visualizando a tabela de clientes

```python
# Definindo a consulta SQL
query = """
SELECT *
FROM `dados-400119.cats_bet.clientes`
LIMIT 10
"""

# Executando a consulta e convertendo para um dataframe do Pandas
df_clientes = client.query(query).to_dataframe()

# Visualize os dados
df_clientes.head()
```

# **Visualizando a Tabela transações**

```python
# Definindo a consulta SQL
query = """
SELECT *
FROM `dados-400119.cats_bet.transacoes`
LIMIT 10
"""
# Executando a consulta e convertendo para um dataframe do Pandas
df_transacoes = client.query(query).to_dataframe()

# Visualize os dados
df_transacoes.head()
```
